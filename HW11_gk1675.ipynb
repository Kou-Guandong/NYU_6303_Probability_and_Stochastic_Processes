## (a)

As $X(t)$ is a Gaussian process, then by definition, 
$X_1 = X(t_1), X_2 = X(t_2),...X_n = X(t_n)$ are jointly Gaussian random variables whose jointly characteristic function is given by: 
$$
\phi_X(\omega_1, \omega_2,..., \omega_n) = exp[j\sum_{k=1}^n \mu(t_k)\omega_k - \frac{1}{2}\sum_{i,k}\sum^n C_{XX}(t_i, t_k)\omega_i \omega_k] \tag{1-1}
$$
where $ C_{XX}(t_i, t_k) = R_{XX}(t_1, t_2) - \mu_x(t_1)\mu_X^* (t_2) $

For a Wide-Sense-Stationary process,
$$
E[X(t)] = \mu \\
E[X(t_1) X^*(t_2)] = R_{XX}(t_2 - t_1)
$$

Insert into (1-1) and we can get
$$
\phi_X(\omega_1, \omega_2,..., \omega_n) = exp[j\sum_{k=1}^n \mu \omega_k - \frac{1}{2}\sum_{i=1}^n\sum_{k=1}^n C_{XX}(t_i-t_k)\omega_i \omega_k] \tag{1-2}
$$
which is only a function of $t_i - t_k$.   
Hence if a new set of time indices are shifted by a contant $c$ to generate a new set of jointly Gaussian random variables $X_1' = X(t_1+c), X_2' = X(t_2+c),...X_n' = X(t_n+c)$, then their characteristic function is identical to (1-2). Thus $\{X_i\}_{i=1}^n$ and $\{X_i'\}_{i=1}^n$ have the same joint probability distribution for all n and c, establishing the strict sense stationary property.

## (b)

If the input $x(t)$ to a memoryless system is WSS then theoutput process may not be stationary in any sense.   
***proof***   
let $g(\cdot)$ denotes a memoryless system, and $y(t) = g[x(t)]$.  
Then $y(t_i)$ only depends on $x(t_i)$. 
Without loss of generality, let 
$$
y(t_1) = g[x(t_1)], y(t_2) = g[x(t_2)], \dots, y(t_n) = g[x(t_n)]
$$
Then the joint probability density function
$$
f_Y(y_1, y_2, \dots, y_n; t_1, t_2, \dots, t_n) = 
\frac{f_X(x_1, x_2, \dots, x_n; t_1, t_2, \dots, t_n)}
{|g^{'}(x_1)g^{'}(x_2)\dots g^{'}(x_n)|}
$$

The denominator $|g^{'}(x_1)g^{'}(x_2)\dots g^{'}(x_n)|$ does not depend on $t$. 
For W.S.S process $X(t)$ in general,
$$
f_X(x_1, x_2, \dots, x_n; t_1, t_2, \dots, t_n) \neq f_X(x_1^{'}, x_2^{'}, \dots, x_n^{'}; t_1+c, t_2+c, \dots, t_n+c) \\
\Rightarrow 
f_Y(y_1, y_2, \dots, y_n; t_1, t_2, \dots, t_n) \neq f_Y(y_1^{'}, y_2^{'}, \dots, y_n^{'}; t_1+c, t_2+c, \dots, t_n+c)
$$
Thus, $y(t)$ may not be stationary in any sense.

***Autocorrelation function***   
$$
R_{XX}(t_1,t_2) = E[X(t_1)X^*(t_2)] \\
= E[cos(t_1 + W(t_1))cos(t_2 + W(t_2))] \\
= E\{Re[e^{j(\omega_0t_1 +W(t_1))}e^{-j(\omega_0 t_2 + W(t_2))}]\} \\

$$

As $W(t)$ is a wiener process,
$$
R_{WW}(t_1, t_2) = min(t_1, t_2) \\

\Rightarrow

t_2 - t_1  & (t_1 \lt t_2) \\
t_1 - t_2 & (t_2 \lt t_1)

where $\tau = t_1 - t_2$
