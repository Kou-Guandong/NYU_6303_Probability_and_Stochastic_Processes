\mu_n = E[X_n] = E\{E[X_n|X_{n-1}]\} = E[\lambda X_{n-1}] = \lambda E[X_{n-1}]

$$

$$

= \lambda^2 \sigma_{n-1}^2 + \lambda^{n-1} \\

= \lambda^{n-1}\frac{\lambda^{n+1}-1}{\lambda-1}

$$

(HY - X) Y^T = 0 \\
HYY^T - XY^T = 0 \\
HR_{YY} - R_{XY} = 0 \\
H = R^{-1}_{YY}R_{XY}
$$

$$
R_{YY} = E[YY^T] = E[
\begin{bmatrix}
X+n \\
X+W

\begin{bmatrix}
X+n & X+w

$$
R^{-1}_{YY} = \frac{1}{\sigma_X^2(\sigma_n^2 + \sigma_w^2) + \sigma_n^2 \sigma_w^2}

$$

$$

\begin{bmatrix}
\sigma_X^2 + \sigma_w^2 & -\sigma_X^2 \\
-\sigma_X^2 & \sigma_X^2 + \sigma_n^2
\end{bmatrix}
\begin{bmatrix}
\sigma_x^2 \\ \sigma_x^2 
\end{bmatrix}

\begin{bmatrix}
\sigma_w^2 \\ \sigma_n^2 
\end{bmatrix}
$$

Mean square error (MSE): 
$$
\epsilon^2 = ||\hat{X} - X ||^2 = E[(HY-X)(HY-X)^T] = E[XX^T - HYX^T - XY^TH^T + HYY^TH^T] \\
= R_{XX} - HR^T_{XY} - R_{XY}H^T + HR_{YY}H^T
$$

When $H=R^{-1}_{YY}R_{XY}$, the MSE has minimum value
$$
\epsilon^2 = R_{XY} - R_{XY}R^{-1}_{YY}R_{XY}^T 
= \frac{\sigma_X^2\sigma_w^2\sigma_n^2}{\sigma_X^2(\sigma_n^2 + \sigma_w^2) + \sigma_n^2 \sigma_w^2}
