# Problem 1

## 1(a)

$$
R_{XX} = E[X(n_1)X^*(n_2)] \\
R_{YY}(n_1, n_2) = E[Y(n_1)Y^*(n_2)] = E[X(pn_1)X^*(pn_2)] = R_{XX}(pn) = \rho^{p|n|} \geq 0 
$$

As $X(t)$ is W.S.S, $\Rightarrow E[X(n)] = const $   
$ \Rightarrow E[Y(n)] = E[X(nt)] = const $  
$\Rightarrow Y(n) $ is W.S.S

$$
S_{YY}(\omega) = \sum_{n=-\infty}^{\infty} \rho^{p|n|}e^{-j\omega n} \\
= \sum_{n=-\infty}^{-1} \rho^{-pn}e^{-j\omega n} + \sum_{n=0}^{\infty} \rho^{pn}e^{-j\omega n} \\
= \sum_{n=1}^{\infty} (\rho^{p}e^{j\omega})^n + \sum_{n=0}^{\infty} (\rho^{p}e^{-j\omega })^n \\
= \frac{\rho^{p}e^{j\omega}}{1- \rho^{p}e^{j\omega} }  +  \frac{1}{1 - \rho^p e^{-j\omega}} \\
= \frac{1-\rho^{2p}}{1 - \rho^p(e^{j\omega + e^{-j\omega}}) + \rho^{2p}} \\
= \frac{1-\rho^{2p}}{1 - \rho^p \cdot 2cos(\omega) + \rho^{2p}} \\
= \frac{\frac{\rho^{2p}-1}{2\rho^2}}{-\frac{1+\rho^{2p}}{2\rho^p} + cos\omega} \\
= \frac{A}{B+cos\omega} \\
\Rightarrow A = \frac{\rho^{2p}-1}{2\rho^p}, B = -\frac{1+\rho^{2p}}{2\rho^p}
$$

## 1(b)

For $R_{XX}(\tau) $ to represent an autocorrelation function, it has to be nonnegative.

$$
R_{XX}(\tau) = e^{-|\tau|} sin(|\tau|+\psi) \geq 0 \\
e^{-|\tau|} > 0 \\
\Rightarrow  sin(|\tau|+\psi)  \geq 0 \\
\Rightarrow |\tau|+\psi \in [2k\pi, (2k+1)\pi], k \in Z \\
\Rightarrow \psi \in [2k\pi - |\tau|, (2k+1)\pi - |\tau|], k \in Z 
$$

Power Spectrual Density:
$$
S_{XX}(\omega) = \int_{-\infty}^{\infty} R_{XX}(\tau) e^{-j\omega \tau}d\tau 
$$

As $R_{XX}(\tau) $ is an even function,
$$
S_{XX}(\omega) = 2\int_{0}^{\infty} R_{XX}(\tau) cos(\omega \tau)d\tau, (\tau \geq 0) \\
= 2\int_{0}^{\infty} e^{-\tau} sin(\tau+\psi)   cos(\omega \tau) d\tau \\
= \int_{0}^{\infty} e^{-\tau} [sin(1+\omega)\tau+\psi) +  sin(1-\omega)\tau+\psi)] d\tau \\
= \frac{(2+\omega^2)sin(\psi) + (2 - \omega^2)cos(\psi)}{\omega^4+4} 
\geq 0, \forall \omega \in [-\infty, \infty]
$$

$$
\Rightarrow
S_{XX}(\omega) = \sqrt{\frac{2}{4+\omega^4}} sin(\psi + \theta), \\
\text{where } \theta = tan^{-1}(\frac{2-\omega^2}{2+\omega^2}) \in [-\frac{\pi}{4}, \frac{\pi}{4}]
$$

$$
S_{XX}(\omega) \geq 0
\Rightarrow \theta +\psi \in [2k\pi, (2k+1)\pi], k \in Z, \theta \in [-\frac{\pi}{4}, \frac{\pi}{4}] \\
\Rightarrow \psi \in [(2k+\frac{1}{4})\pi, (2k+\frac{3}{4})\pi], k \in Z
$$

$$

$$
R_{YY}(t_1, t_2) = E[Y(t_1)Y^*(t_2)] = E[e^{jX(t_1)}e^{-jX(t_2)}] = E[e^{j[X(t_1)-X(t_2)]}]

$$
\Rightarrow R_{YY}(t_1,t_2) = E[e^{jX(\tau)}] = e^{-\frac{|\tau|}{2}}, \tau=t_1 -t_2

$$
S_{XX}(\omega) = \int_{-\infty}^{\infty} e^{-\alpha|\tau|}cos(\omega_0 \tau) e^{-j\omega\tau} d\tau
$$

As $e^{-\alpha|\tau|}cos(\omega_0 \tau)$ is even function, 
$$
S_{XX}(\omega) = 2\int_{0}^{\infty} e^{-\alpha\tau} cos(\omega_0 \tau) cos(\omega\tau) d\tau \\
$$

Seperate $cos(\omega_0 \tau)$ from the integral,
$$
2\int_{0}^{\infty} e^{-\alpha\tau} cos(\omega\tau) d\tau \\
= \int_{0}^{\infty} [e^{(j\omega-a)\tau} + e^{-(j\omega + a)\tau}]d\tau \\
= \frac{1}{j\omega - \alpha} - \frac{1}{\alpha + j\omega} \Bigg\vert_{\tau=0}^{\infty} \\
= \frac{2\alpha}{\alpha^2 + \omega^2}
$$

With the frequency shift property of Fourier transform,
$$
S_{XX}(\omega) = 2\int_{0}^{\infty} e^{-\alpha\tau} cos(\omega_0 \tau)  cos(\omega\tau) d\tau \\
= \frac{\alpha}{\alpha^2 + (\omega + \omega_0)^2} + \frac{\alpha}{\alpha^2 + (\omega - \omega_0)^2}
$$

$$
S_{XX}(\omega) = \int_{-\infty}^{\infty} e^{-\alpha\tau^2} cos(\omega_0 \tau) e^{-j\omega\tau} d\tau
$$

With the property of [Gaussian integral](https://en.wikipedia.org/wiki/Gaussian_integral#Generalizations)

$$
\int_{-\infty}^{\infty} e^{-\alpha\tau^2} e^{-j\omega\tau} d\tau \\
= \int_{-\infty}^{\infty} e^{-\alpha(\tau^2 + \frac{j\omega\tau}{\alpha} - \frac{\omega^2}{4\alpha^2})

= e^{\frac{-\omega^2}{4\alpha}} \int_{-\infty}^{\infty} e^{-\alpha(x+b)^2}dx, (x = \tau + \frac{j\omega}{2\alpha}) \\
= \sqrt{\frac{\pi}{\alpha}}e^{-\frac{\omega^2}{4\alpha}}
$$

With the frequency shift property of Fourier transform,
$$
